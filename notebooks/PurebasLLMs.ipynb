# Pruebas con el modelo llama-3.3-70b-versatile
Cargamos la contraseña de la clave de y le pasaremos un promt sencillo para ver si podemos hacer que funcione bien.

```python

load_dotenv()
print(os.environ["GROQ_API_KEY"][:12])


PROVIDER = "groq"
groq_model = "llama-3.3-70b-versatile" 
llm_client = Groq()
model = groq_model
Especie = "Galerina marginata"


prompts = [{"role": "system", "content":'Eres un experto en alimentación humana, que tiene que dar inforamción sobre setas'},
            {"role": "system", "content":'Tienes que determinar que seta es comestible o no y porfavor se breve.'},
            {"role": "system", "content":'En el caso de ser comestible da sugerencias de cocina.'},
            {"role": "system", "content":'En caso de no ser comestible da razones de porque no consumirla'},
            {"role": "system", "content":'Sigue la siguiente estructura: "Nombre: Aquí va el nombre de la seta\n Comestible: Si/NO\n Recetas de cocida/Razones", las razones/recetas son una lista'},
        {"role": "user", "content":f'¿Qué puedes decirme sobre {Especie}?'}]
respuesta_llm = llm_client.chat.completions.create(
    model=model,
    messages=prompts,
    temperature=0,
    max_tokens=1000
)

contenido_respuesta = respuesta_llm.choices[0].message.content

print(contenido_respuesta)
```

Como nuestro modelo de clasificación inicial nos permite saber si la especie es venenosa o no, lo mejor que podemos hacer sería realizar prompts diferentes segun el caso.

```python
info = "no se sabe" #comestible, no comestible o no se sabe
Especie = "phaeophyscia orbicularis"

prompts = [{"role": "system", "content":'Eres un experto en alimentación humana, que tiene que dar inforamción sobre setas'},
           {"role": "system", "content": 'No es necesario que seas muy técnico explica todo de forma sencilla para todos'}]
if info == "comestible":
    prompts+=[{"role": "system", "content":'Se te pasará una seta que se sabe que es comestible, queremos que nos des recetas de cocina'},
            {"role": "system", "content":'Se breve'},
            {"role": "system", "content":'Sigue la siguiente estructura: "Nombre: Aquí va el nombre de la seta\n Comestible: Si/NO\n Recetas de cocida", las recetas son una lista'},
            {"role": "user", "content":f'¿Qué puedes decirme sobre {Especie}?'}]
elif info == "no comestible":
    prompts+=[{"role": "system", "content":'Se te pasará una seta que se sabe que es no comestible, queremos que nos des recetas de cocina'},
            {"role": "system", "content":'Se breve'},
            {"role": "system", "content":'Sigue la siguiente estructura: "Nombre: Aquí va el nombre de la seta\n Comestible: Si/NO\n Razones para no consumirla", las razones dala como una lista '},
            {"role": "user", "content":f'¿Qué puedes decirme sobre {Especie}?'}]
elif info == "no se sabe":
    prompts+=[{"role": "system", "content":'En este caso se te dará una especie que no se sabe si es comestrible o no'},
            {"role": "system", "content":'Esta categoría se asigna si no hay valor culinario, no es una seta, varía según lo madura que este o incluso porque no existe página wikipedia de la misma'},
            {"role": "system", "content":'Se breve y habla de la misma según cada caso'},
            {"role": "user", "content":f'¿Qué puedes decirme sobre {Especie}?'}]
else:
    prompts+=[{"role": "system", "content":'En este caso no entra en ninguna categoría comestible o no comestible'},
            {"role": "user", "content":'Vale con que digas que no se sabe'}]
    
respuesta_llm = llm_client.chat.completions.create(
    model=model,
    messages=prompts,
    temperature=0.2,
    max_tokens=1000
)

contenido_respuesta = respuesta_llm.choices[0].message.content
print(contenido_respuesta)
```
